#### Check and Validate Similarity based Recommendation

We used cosine similarity to find the most similar handbags based on input vector. <br><br>
To analyze the similary between handbags:
1. Use handbag vector from vectors_test as input
2. Measure sim1 and sim2 which are the similarity between input vectors and handbag vectors from my vectors_shein and vectors_saks. 
3. Measure sim3, the similarity between input vectors and handbag vectors from my vectors_test with same handbag name, which means they are the different images for sim handbag.
4. My reconmmendatation is based on sim1 and sim2.
5. If the sim3 can significantly higher than sim1/sim2, that means to measure the similarity between my generated handbag vectors is able to determine the handbag similarity, which also validates my recommendation system.

```python
def provide_sim(input_vectors, vectors_shein, vectors_shop):
    """
    Calculate similarity between input_vectors and vectors_shein, and input_vectors and vectors_shop
    
    """
    sim1 = cosine_similarity(input_vectors, vectors_reorder)
    sim1 = np.reshape(sim1, np.shape(sim1)[1])
    sim1_dict = [["sim1_" + str(i),j] for (i,j) in zip(range(len(sim1)), sim1)]
    sim2 = cosine_similarity(input_vectors, vectors_shein)
    sim2 = np.reshape(sim2, np.shape(sim2)[1])
    sim2_dict = [["sim2_" + str(i),j] for (i,j) in zip(range(len(sim2)), sim2)]
    sim1_dict.extend(sim2_dict)

    return sim1_dict
```

```python
def check_sim(input_vectors, vectors_shein, vectors_shop, other_vectors):
    sim1 = cosine_similarity(input_vectors, vectors_shop)
    sim1 = np.reshape(sim1, np.shape(sim1)[1])
    sim1_dict = [["sim1_" + str(i),j] for (i,j) in zip(range(len(sim1)), sim1)]
    sim2 = cosine_similarity(input_vectors, vectors_shein)
    sim2 = np.reshape(sim2, np.shape(sim2)[1])
    sim2_dict = [["sim2_" + str(i),j] for (i,j) in zip(range(len(sim2)), sim2)]
    sim3 = cosine_similarity(input_vectors, other_vectors)
    sim3 = np.reshape(sim3, np.shape(sim3)[1])
    sim3_dict = [["sim3_" + str(i),j] for (i,j) in zip(range(len(sim3)), sim3)]
    
    sim1_dict.extend(sim2_dict)
    sim1_dict.extend(sim3_dict)
    
    return sim1_dict

def validate_sim(input_vectors, vectors_shein, vectors_shop, other_vectors, sim_type="median"):
    sim_final = check_sim(input_vectors, vectors_shein, vectors_shop, other_vectors)
    sim_data = pd.DataFrame(sim_final)
    sim_data[2] = [ i.split("_")[0] for i in list(sim_data[0])]
    sim_data["order"] = sim_data[1].rank(ascending=False)
    sim_data.columns = ["initial_index", "sim", "index", "order"]
    if sim_type == "median":
        group_data = sim_data.groupby("index").agg({"sim": np.median})
    elif sim_type == "max":
        group_data = sim_data.groupby("index").agg({"sim": np.max})
    print(sim_type + "_sim_saks:" + str(group_data[group_data.index == "sim1"]["sim"].values[0]) )
    print(sim_type + "_sim_shein:" + str(group_data[group_data.index == "sim2"]["sim"].values[0]) )
    print(sim_type + "_sim_othersametest:" + str(group_data[group_data.index == "sim3"]["sim"].values[0]) )
    return sim_data, group_data[group_data.index == "sim1"]["sim"].values[0], group_data[group_data.index == "sim2"]["sim"].values[0], group_data[group_data.index == "sim3"]["sim"].values[0]

def get_test_vectors(test_vectors, bag_type, total_imgs):
    bag_index = [idx for idx, i in enumerate(total_imgs) if i.split("/")[-2] == bag_type]
    test_index = random.sample(bag_index, 1)[0]
    print("sampled index:", test_index)
    test_vector = test_vectors[test_index:test_index + 1]
    other_vectors = [test_vectors[i] for i in bag_index if i != test_index]
    return test_index, [ i for i in bag_index if i != test_index], test_vector, other_vectors
```

### Handbag Recommendation

Finally, let's try to do some recommendation!

```python
def recommend_bag_sim(input_vectors, vectors_bags):
    sim = cosine_similarity(input_vectors, vectors_bags)
    sim = np.reshape(sim, np.shape(sim)[1])
    sim_dict = [["sim_" + str(i),j] for (i,j) in zip(range(len(sim)), sim)]
    
    sim_data = pd.DataFrame(sim_dict)
    sim_data[2] = [ i.split("_")[0] for i in list(sim_data[0])]
    sim_data["order"] = sim_data[1].rank(ascending=False)
    sim_data.columns = ["initial_index", "sim", "index", "order"]
    return sim_data
```

```python

def recommend_bag(input_vector, shop_data_saks, shop_data_shein, price_range, vetors_shein, vectors_shop):
    price_dict = {"less 50": ["less 50"],
                 "less 250" : ["less 50", "less 250"],
                 "less 500" : ["less 50", "less 250", "less 500"],
                 "less 750" : ["less 50", "less 250", "less 500", "less 750"],
                 "less 1000" : ["less 50", "less 250", "less 500", "less 750", "less 1000"],
                 "higher 1000" : ["higher 1000"]}
    if price_range == "less 50":
        sim_data = recommend_bag_sim(input_vector, vectors_shein)
        shop_data = pd.merge(shop_data_shein, sim_data, on = shop_data_shein.index)
        shop_data.drop("key_0", axis=1, inplace=True)
        shop_data = shop_data[shop_data["price_level"].isin(price_dict[price_range])].sort_values("sim", ascending=False)
        Top5 = shop_data[0:5]["image"].values
    
    elif price_range == "higher 1000":
        sim_data = recommend_bag_sim(input_vector, vectors_shop)
        shop_data = pd.merge(shop_data_saks, sim_data, on = shop_data_saks.index)
        shop_data.drop("key_0", axis=1, inplace=True)
        shop_data = shop_data[shop_data["price_level"].isin(price_dict[price_range])].sort_values("sim", ascending=False)
        Top5 = shop_data[0:5]["image"].values
    
    else:
        sim_data_saks = recommend_bag_sim(input_vector, vectors_shop)
        shop_data_saks = pd.merge(shop_data_saks, sim_data_saks, on = shop_data_saks.index)
        shop_data_saks.drop("key_0", axis=1, inplace=True)
        
        sim_data_shein = recommend_bag_sim(input_vector, vectors_shein)
        shop_data_shein = pd.merge(shop_data_shein, sim_data_shein, on = shop_data_shein.index)
        shop_data_shein.drop("key_0", axis=1, inplace=True)
        
        shop_data = pd.concat([shop_data_saks, shop_data_shein], axis=0)
        shop_data = shop_data[shop_data["price_level"].isin(price_dict[price_range])].sort_values("sim", ascending=False)
        Top5 = shop_data[0:5]["image"].values
    
    return shop_data, Top5
```
